```python
import pandas as pd
import glob
```

```python
glob.glob("tctodd3/*.tsd")[0].split('\\')[1].split('.')[0]
```

```python
filenames = glob.glob("tctodd3/*.tsd")
```

```python
files = [file.split('\\')[1].split('.')[0] for file in filenames]
```

```python
signs = pd.DataFrame()
for filename in files:
    df = pd.read_table('tctodd3/' + filename + '.tsd', header=None)
    label = filename.split('-')[0]
    ID = filename.split('-')[1]
    df['label'] = label
    df['ID'] = ID
    df.columns = ['x_L','y_L','z_L','roll_L','pitch_L','yaw_L','thumb_L','fore_L','middle_L','ring_L','little_L','x_R','y_R','z_R','roll_R','pitch_R','yaw_R','thumb_R','fore_R','middle_R','ring_R','little_R', 'label', 'ID']
    signs = signs.append(df)
```
